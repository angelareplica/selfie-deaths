```python
df.groupby(['Country'])['Casualties'].agg('sum')
```

```python
casualties_sum_df = df.groupby(['Country'])['Casualties'].sum().reset_index()
casualties_sum_df
```

```python
codes_df = pd.read_csv('worldcodes.csv')
codes_df.head()
```

```python
codes_df['Country'] = codes_df['name']
```

```python
merged_df = pd.merge(casualties_sum_df, codes_df, how='left', on='Country')
```

```python
merged_df
```

```python
merged_df = merged_df.drop(columns=['name', 'alpha-2', 'country-code', 'iso_3166-2', 'region', 'sub-region', 'intermediate-region', 'region-code', 'sub-region-code', 'intermediate-region-code'])
```

```python
merged_df['code'] = merged_df['alpha-3']
merged_df = merged_df.drop(columns=['alpha-3'])
merged_df
```

```python
#re-do the above, but keep all the countries
merged_df2 = pd.merge(casualties_sum_df, codes_df, how='right', on='Country')
merged_df2 = merged_df2.drop(columns=['name', 'alpha-2', 'country-code', 'iso_3166-2', 'region', 'sub-region', 'intermediate-region', 'region-code', 'sub-region-code', 'intermediate-region-code'])
merged_df2['code'] = merged_df2['alpha-3']
merged_df2 = merged_df2.drop(columns=['alpha-3'])
# replace all the NaNs with 0
merged_df2.fillna(0)
```

```python
merged_df.to_csv("selfiedeaths_grouped_coded.csv", index=False)
```
